```python
quantiles = pd.qcut(df[:1395]['date'], q=8, labels=False, duplicates='drop')

# Step 2: Use pd.get_dummies to convert the numerical bins into one-hot encoded columns
# `prefix='Chunk'` names the new columns nicely (e.g., 'Chunk_0', 'Chunk_1', etc.)
chunk_dummies = pd.get_dummies(quantiles, prefix='t', dtype=int)

# Step 3: Concatenate the new columns back to the original DataFrame
df = pd.concat([df, chunk_dummies], axis=1)
```

```python
df.fillna(0, inplace=True)
```
